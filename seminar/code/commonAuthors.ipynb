```python
# Used libraries
import codecs
import json
import math
from collections import Counter

import nltk

nltk.download('vader_lexicon')
nltk.download('stopwords')

import warnings
```

```python
# There are some warnings with LDA which we want to ignore
warnings.filterwarnings('ignore')
```

```python
def load_tweets(filenames):
    """
    Loads the tweets from the file with the given name into an array of tweets.

    @param filename: The filename of the file to load the tweets from.

    @returns: An array of tweets.
    """
    tweets = []
    for filename in filenames:
        with open(filename, 'r') as f:
            for sLine in f:
                tweet = json.loads(sLine)
                tweets.append(tweet)
    return tweets
```

```python
# Name of the json file with the tweets to process
tweets_filename = ['tennis_combined_filtered2_4_days.json', "table_tennis_filtered2_4_days.json", "badminton_filtered2_4_days.json"]

# File to store the most common user ids to
filename = "most_common_user_ids.json"

# number of tweets to be relevant:
tweet_tresh = 5

author_tresh = 10000
```

```python
# Load the tweets
tweets = load_tweets(tweets_filename)
print(len(tweets))
```

```python
# Use a counter for counting hashtags
author_counter = Counter()

# Add hashtags to counter
author_ids = []
for tweet in tweets:
    author_id = tweet.get("author_id")
    if author_id is not None:
        author_ids.append(author_id)

author_counter.update(author_ids)

# Print most used hashtags
for author_id, count in author_counter.most_common(author_tresh):
    print(str(author_id) + ": " + str(count))
```

```python
# Store user ids
with open(filename, 'w') as json_file:
    for author_id, count in author_counter.most_common(author_tresh):
        if count > tweet_tresh:
            json.dump(author_id, json_file)
            json_file.write('\n')

print("User ids successfully stored to: ", filename)
```
